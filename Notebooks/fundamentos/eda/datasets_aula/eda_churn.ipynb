# Analise Bivariada

```python
# Hipotese: Cliente com contrato do tipo mensal esta mais propenso ao churn
# Contruir uma tabela de contingencia entre tipo mensal e churn - quantidade
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total') 
```

```python
# Contruir uma tabela de contingencia entre tipo mensal e churn - proporção
pd.crosstab(df_churn.Churn, df_churn.Contract, normalize='index', margins=True, margins_name='Total') 
```

Considerando o public total, os clientes de contrato mensal representam 55%. Porém , se consideramros apenas os clientes que abandonaram o serviço, 88%% tinham contrato mensal. Desta forma, podemos afirmar que há uma correlação entre o tipo de contrato mensal e o abandono

```python
# Avaliar a correlção entre duas variaveis categoricas(qualitativas) 

# Executar um teste de hipotese qui-quadrado de pearson
# Num teste de hipotes, duas hipoteses sao formuladas:
# H0: hipotese nula, as duas variavaeis são independentes
# H1: hipotese complementar, as duas variaveis nçao sao independentes

# O teste serve para confirmar ou recusar a H0
# Quando a probabilidade observamos H0 é inferior a 0.05(p-value), recusamos H0 e seguimos com H1
```

```python
# Gerar um DF da crosstab(sem totais)
df_crosstab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
# Calcular os Scores e p-values
from scipy.stats import chi2_contingency
```

```python
# Tupla = lista imutavel
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# Apresentar numeros com decimais sem notacao cientifica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
# Criar um DataFrame
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': scores_churn_contract, 'p-value': pvalues_churn_contract})
```

```python
# Analisar scores e pvalues
df_chi_scores_churn_contract
```

Uma vez que p-value é menor que 0.05 recusamos a H0, logo as variaveis não sao independentes. Pelo qui2 alto, podemos afirmar que há uma forte correlação

```python
# Analisar a correlaçao entre uma variavel qualitativa e uma quantitativa
# Hipotese: cliente com menos de 6 meses de contrato é propenso ao churn

# Criar uma coluna nova do dataFrame com base na condição de outra coluna
# Coluna = 'TempoMenor6meses'

import numpy as np
```

```python
# Metodo de Tukey - IQR(Distribuições qaue não seguem a distribuição normal)
# IQR - Range Interquartil
# IQR = 3°quartil - 1°quartil

q1_totalCharges_month = df_churn_month.TotalCharges.quantile(0.25)
q3_totalCharges_month = df_churn_month.TotalCharges.quantile(0.75)
iqr_totalCharges_month = q3_totalCharges_month-q1_totalCharges_month
```

```python
iqr_totalCharges_month
```

```python
# Limites inferior e superior
limInf_TotalCharge_month = q1_totalCharges_month - (iqr_totalCharges_month * 1.5)
limSup_TotalCharge_month = q3_totalCharges_month + (iqr_totalCharges_month * 1.5)
limInf_TotalCharge_month, limSup_TotalCharge_month
```

```python
df_churn_month[(df_churn_month.TotalCharges < limInf_TotalCharge_month) | (df_churn_month.TotalCharges > limSup_TotalCharge_month)]
```

```python
# ZScore (Usado para a distribuição Normal)
# Indica quantos desvios padrões um ponto especifico de dados esta distante da media
# zcore = (x-media) / desvioPadrao
from scipy.stats import zscore
```

```python
z = np.abs(zscore(df_churn_month.TotalCharges))
```

```python
z
```

```python
df_churn_month[z>3.0]
```
